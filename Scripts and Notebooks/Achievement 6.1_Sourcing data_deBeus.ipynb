### Now let's create some categorical variables based on rates

```python
countyHealthFactors.describe()
```

```python
#let's create categories based on national median (since housing burden mean is high because of an outlier)
countyHealthFactors['alcoholInvolvementDrivingDeathsRate'].value_counts(dropna = False)
```

```python
countyHealthFactors.loc[countyHealthFactors['alcoholInvolvementDrivingDeathsRate'] == 0.00, 'alcoholDrivingDeathsRateCategory'] = 0
countyHealthFactors.loc[(countyHealthFactors['alcoholInvolvementDrivingDeathsRate'] > 0) & (countyHealthFactors['alcoholInvolvementDrivingDeathsRate'] <= 25.6), 'alcoholDrivingDeathsRateCategory'] = 1
countyHealthFactors.loc[countyHealthFactors['alcoholInvolvementDrivingDeathsRate'] > 25.6, 'alcoholDrivingDeathsRateCategory'] = 2
```

```python
#let's do zero deaths, less than median, above median
countyHealthFactors['alcoholDrivingDeathsRateCategory'].value_counts()
```

```python
countyHealthFactors['severeHousingProblemsRate'].value_counts()
```

```python
countyHealthFactors.loc[countyHealthFactors['severeHousingProblemsRate'] <= 12.6, 'housingProblemsRateCategory'] = 1
countyHealthFactors.loc[countyHealthFactors['severeHousingProblemsRate'] > 12.6, 'housingProblemsRateCategory'] = 2
```

```python
countyHealthFactors['housingProblemsRateCategory'].value_counts()
```

```python
countyHealthFactors['householdsWithBroadbandAccessRate'].value_counts()
```

```python
#below median
countyHealthFactors.loc[countyHealthFactors['householdsWithBroadbandAccessRate'] <= 81.7, 'broadbandAccessRateCategory'] = 1
#above median
countyHealthFactors.loc[countyHealthFactors['householdsWithBroadbandAccessRate'] > 81.7, 'broadbandAccessRateCategory'] = 2
```

```python
countyHealthFactors['broadbandAccessRateCategory'].value_counts()
```
